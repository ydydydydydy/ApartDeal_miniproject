```python
### merged를 사용한다면 법정동의 여러 타입을 한번에 할수 없음. (리, ~가, 등)
# for loop 사용해서 하나씩 넣기
# '시군구명':'지역이름', '리명':'법정동'
pop_list = []
for i in range(len(sidoGuDong)) :
    if sidoGuDong['법정동'][i][-1] == ('동' or '가')  :
        
        try:
            temp = int(dongIngu[ (dongIngu['시군구명']==sidoGuDong['지역이름'][i]) & (dongIngu['리명']==sidoGuDong['법정동'][i]) ]['전체 당월인구수'].values[0])
            pop_list.append(temp)
        except :
            pop_list.append(-40)

    
    ## 리는 면/읍 과 합쳐져서 법정동에 들어가있음. 
    elif sidoGuDong['법정동'][i][-1] == '리'  :
        
        if " " in sidoGuDong['법정동'][i] :
        
            ind = sidoGuDong['법정동'][i].index(" ")       

        try :
            temp = int(dongIngu[ (dongIngu['시군구명']==sidoGuDong['지역이름'][i]) & (dongIngu['리명']== sidoGuDong['법정동'][i][ind+1:])  ]['전체 당월인구수'].values[0])
            pop_list.append(temp)

        except : 
            pop_list.append(-60)

        
    ## 데이터 매칭실패. -100 넣어놓고 후에 전처리
    else :
        pop_list.append(-100)
        
        
```

## 시도 + 법정동 groupby. + 인구수

```python
data
```

```python
sidoDong = data.groupby(['시도', '법정동'], as_index=False).agg(평균거래금액=('거래금액', 'mean'), 
                                                           거래수=('거래금액', 'count'),
                                                          평균전용면적 = ('전용면적','mean'))
sidoDong
```

```python
### for loop 사용해서 인구수 추가

# '시도명':'시도', '리명':'법정동'
pop_list = []
for i in range(len(sidoDong)) :
    
#     '~동' , '~가'
    if (sidoDong['법정동'][i][-1] == '동') or (sidoDong['법정동'][i][-1] == '가'):
        
        try:
            temp = int(dongIngu[ (dongIngu['시도명']==sidoDong['시도'][i]) & (dongIngu['리명']==sidoDong['법정동'][i]) ]['전체 당월인구수'].values[0])
            pop_list.append(temp)
        except :
            pop_list.append(-40)


            
#     if   :
        
#         try:
#             temp = int(dongIngu[ (dongIngu['시도명']==sidoDong['시도'][i]) & (dongIngu['리명']==sidoDong['법정동'][i]) ]['전체 당월인구수'].values[0])
#             pop_list.append(temp)
#         except :
#             pop_list.append(-40)
            
            
            
    
    ## 리는 면/읍 과 합쳐져서 법정동에 들어가있음. 
    elif sidoDong['법정동'][i][-1] == '리'  :
        
        try :
            if " " in sidoDong['법정동'][i] :
                ind = sidoDong['법정동'][i].index(" ")       
                temp = int(dongIngu[ (dongIngu['시도명']==sidoDong['시도'][i]) & (dongIngu['리명']== sidoDong['법정동'][i][ind+1:])  ]['전체 당월인구수'].values[0])
            else :
                temp = int(dongIngu[ (dongIngu['시도명']==sidoDong['시도'][i]) & (dongIngu['리명']== sidoDong['법정동'][i])  ]['전체 당월인구수'].values[0])
            
            pop_list.append(temp)
            
        except : 
            pop_list.append(-60)

        
    ## 데이터 매칭실패. -100 넣어놓고 후에 전처리
    else :
        pop_list.append(-100)
```

```python
## 인구 컬럼 추가
sidoDong['인구수'] = pop_list
sidoDong
```
